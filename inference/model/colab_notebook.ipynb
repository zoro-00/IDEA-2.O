## 4. Setup Paths (`data_config.json`)
Your local `data_config.json` contains Windows absolute paths. This script will automatically update it to match the Colab Google Drive environment.

```python
import json
import os

# Create data and models directories if they don't exist
os.makedirs(f"{PROJECT_DIR}/data", exist_ok=True)
os.makedirs(f"{PROJECT_DIR}/models", exist_ok=True)

config = {
    "paths": {
      "aml_data": f"{PROJECT_DIR}/data", 
      "model_to_load": f"{PROJECT_DIR}/models",
      "model_to_save": f"{PROJECT_DIR}/models"
    }
}

with open('data_config.json', 'w') as f:
    json.dump(config, f, indent=4)

print("Updated data_config.json to use Drive paths:")
!cat data_config.json
```
